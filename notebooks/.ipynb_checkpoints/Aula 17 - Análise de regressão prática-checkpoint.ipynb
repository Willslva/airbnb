 # Mini-dicionário de dados

* **Customer Information:**
    * Customer ID: A unique identifier for each customer.
    * Name: Customer’s full name.
    * Email: Customer’s email address for communication.
    * Phone: Contact number of the customer.
    * Address: Physical address of the customer.
    * City, State, Zipcode, Country: Geographic details of the customer.
    * Age: Age of the customer.
    * Gender: Gender of the customer.
    * Income: Income bracket or level of the customer.
    * Customer Segment: Classification of customers based on behavior or demographics. 

* **Transaction Details:**
    * Last Purchase Date: Date of the customer’s most recent purchase.
    * Total Purchases: Total number of purchases made by the customer.
    * Amount Spent: Total monetary amount spent by the customer.

* **Product Information:**
    * Product Category: Category to which the purchased product belongs (e.g., electronics, clothing, groceries).
    * Product Brand: Brand name of the product.
    * Product Type: Type or model of the product purchased.

* **Feedback:**
    * Feedback: Customer’s feedback or rating related to the product or service received.

* **Transaction Logistics:**
    * Shipping Method: The method used to deliver the purchased products.
    * Payment Method: The method of payment chosen by the customer.
    * Order Status: Status of the order (e.g., shipped, delivered, canceled).

# Feature engineering

**Features criadas**
 
* Dia da compra -> `date`
* Horário para segundos -> `time`
* Parte do dia (manhã, tarde, etc) -> `time`
* city_income -> `city` + `income`
* customer_segment_income -> `customer_segment` + `income`
* gender_income -> `gender` + `income`
* customer_segment_month -> `customer_segment` + `month`
* product_category_month -> `product_category` + `month`
* product_category_customer_segment -> `product_category` + `customer`

```python
join_textual_variables = lambda x: (x.iloc[0] + '_' + x.iloc[1]).lower()
```

```python
retail.loc[:, 'city_income'] = retail.loc[:, ['city', 'income']].apply(
    join_textual_variables, axis = 1
)

retail.loc[:, 'customer_segment_income'] = retail.loc[:, ['customer_segment', 'income']].apply(
    join_textual_variables, axis = 1
)

retail.loc[:, 'gender_income'] = retail.loc[:, ['gender', 'income']].apply(
    join_textual_variables, axis = 1
)

retail.loc[:, 'customer_segment_month'] = retail.loc[:, ['customer_segment', 'month']].apply(
    join_textual_variables, axis = 1
)

retail.loc[:, 'product_category_month'] = retail.loc[:, ['product_category', 'month']].apply(
    join_textual_variables, axis = 1
)

retail.loc[:, 'product_category_customer_segment'] = retail.loc[:, ['product_category', 'customer_segment']]\
.apply(join_textual_variables, axis = 1)
```

```python
# time features
convert_to_seconds = lambda x: int(x[:2]) * 3600 + int(x[3:5]) * 60 + int(x[6:])
retail['time_seconds'] = retail['time'].apply(lambda x: '0' * (8 - len(x)) + x if len(x) < 8 else x)
retail['time_seconds'] = retail['time_seconds'].apply(convert_to_seconds)
```

```python
from datetime import datetime

def correct_date(date):
    date_splitted = date.split('-')
    if len(date_splitted[-1]) < 4:
        results = date_splitted[:2] + ['20' + date_splitted[-1]]
        results = ''.join(results)
    elif len(date_splitted[0]) < 2:
        results = ['0' + date_splitted[0]] + date_splitted[1:]
        results = ''.join(results)
    else:
        results = date.replace('-', '')
        
    return datetime.strptime(results, '%m%d%Y')
```

```python
retail['date'] = retail['date']\
.str.replace('/', '-').apply(correct_date)
```

```python
# date features

retail['day'] = retail['date'].dt.day

us_holidays = holidays.country_holidays('US')

retail['is_holiday'] = retail['date']\
.apply(lambda x: us_holidays.get(x) is not None).astype(int)

retail['day_of_week'] = retail['date'].dt.dayofweek
retail['is_weekend'] = (retail['date'].dt.dayofweek > 4)\
.astype(int)

retail['quarter'] = retail['date'].dt.quarter
retail['last_days_month'] = (retail['day'] >= 28).astype(int)
```

#### Estimação da distribuição da variável resposta (`total_amount`) e análise de outliers

Por se tratar de valores monetário gastos na plataforma, sem devoluções, esperamos que a distribuição de probabilidade seja positiva. Esta hipótese se confirma através da estimação da distribuição de probabilidade via distância de Wasserstein e teste de hipótese de Kolmogorov-Smirnov (KS).

Nesse caso, temos que a melhor distribuição que se ajusta aos dados é uma Gamma com location = 9.99 e scale = 1119, como pode ser visto no gráfico abaixo. Outras distribuições também podem ser consideradas, a saber Exponencial, Beta e Pareto.

```python
dfit = distfit(alpha=.05, distr='popular', random_state = 42, verbose = False)
dfit.fit_transform(retail['total_amount'], verbose = False)
```

##### Outliers na variável resposta

Analisando exclusivamente a variável resposta (`total_amout`), temos aproximadamente 1.30% de outliers o que corresponde a 3816 transações. Contudo, não faremos nenhuma ação sobre os outliers, primeiramente iremos aplicar um modelo de regressão para entender a capacidade do mesmo entender esses valores discrepantes. A depender do tipo de efeito sob o modelo/algoritmo, podemos separar, excluir, melhorar o ajuste do modelo ou selecionar outro mais robusto.

```python
sns.boxplot(x = retail['total_amount']);
```

```python
def detect_outliers(series):
    # Calculate Quartiles and IQR
    q1 = series.quantile(.25)
    q3 = series.quantile(.75)
    iqr = q3 - q1
    
    # Set a threshold for anomaly detection (e.g., 1.5 times IQR)
    threshold_lower = q1 - 1.5 * iqr
    threshold_upper = q3 + 1.5 * iqr
    
    # Identify anomalies
    outliers = series[(series < threshold_lower) | (series > threshold_upper)]
    return outliers, threshold_lower, threshold_upper
```

```python
outlier_values, lower, upper = detect_outliers(retail['total_amount'])
```

```python
outlier_index = outlier_values.index.tolist()
```

```python
len(outlier_index) / retail.shape[0] * 100
```
